```python
def binary_search_min_gamma(f, left=1e-10, right=1e-3, rel_tol=1e-3, max_iter=100,
                            saturation_threshold=1.5, verbose=False):
    """
    Binary search to find gamma where f(gamma) = 0.

    The function f(gamma) = accumulated_error - precision is monotonically increasing
    with gamma, but saturates to ~2 when gamma is large (error saturates to 2).

    Args:
        f: Function that returns (error - precision). Should be increasing with gamma.
        left: Initial left bound (small gamma, should give f(left) < 0)
        right: Initial right bound (large gamma, should give f(right) > 0)
        rel_tol: Relative tolerance for convergence (stops when right/left < 1 + rel_tol)
        max_iter: Maximum iterations
        saturation_threshold: Values above this indicate saturation (error ~ 2)
        verbose: Print debug info
    """
    print('===========binary_search_min_gamma===========')

    # Phase 1: Find valid left bound where f(left) < 0
    f_left = f(left)
    expand_count = 0
    while f_left > 0 and expand_count < 20:
        left /= 10
        f_left = f(left)
        if verbose: print(f'shrink left to {left:.2e}, f(left) = {f_left:.4f}')
        expand_count += 1

    if f_left > 0:
        print(f'Warning: Could not find left bound where f(left) < 0. f({left:.2e}) = {f_left:.4f}')
        return None, None

    # Phase 2: Find valid right bound where f(right) > 0 but not saturated
    # Start from a reasonable guess and work outward
    f_right = f(right)
    expand_count = 0

    # If right is saturated (f_right > saturation_threshold), reduce it
    while f_right > saturation_threshold and expand_count < 20:
        right /= 2
        f_right = f(right)
        if verbose: print(f'reduce saturated right to {right:.2e}, f(right) = {f_right:.4f}')
        expand_count += 1

    # If right is negative, increase it
    expand_count = 0
    while f_right < 0 and expand_count < 20:
        right *= 10
        f_right = f(right)
        if verbose: print(f'enlarge right to {right:.2e}, f(right) = {f_right:.4f}')
        expand_count += 1
        # Check for saturation after enlarging
        if f_right > saturation_threshold:
            right /= 5  # Step back a bit
            f_right = f(right)

    if f_right < 0:
        print(f'Warning: Could not find right bound where f(right) > 0. f({right:.2e}) = {f_right:.4f}')
        return None, None

    if verbose:
        print(f'Initial bounds: left={left:.2e} (f={f_left:.4f}), right={right:.2e} (f={f_right:.4f})')

    # Phase 3: Binary search with relative tolerance
    iter_count = 0
    while (right / left > 1 + rel_tol) and iter_count < max_iter:
        mid = np.sqrt(left * right)  # Geometric mean for log-scale search
        f_mid = f(mid)

        if verbose:
            print(f'iter {iter_count}: left={left:.2e}, mid={mid:.2e}, right={right:.2e}, f(mid)={f_mid:.4f}')

        # Handle saturation: if mid gives saturated result, treat as too large
        if f_mid > saturation_threshold:
            right = mid
        elif f_mid < 0:
            left = mid
        else:
            right = mid

        iter_count += 1

    if iter_count == max_iter:
        print(f'Warning: Maximum iterations ({max_iter}) reached.')

    gamma_star = np.sqrt(left * right)  # Geometric mean of final bounds
    print(f'gamma_star = {gamma_star:.6e} (converged in {iter_count} iterations)')

    return gamma_star, f(gamma_star, return_r=True)
```
