```python
import numpy as np
from quantum_simulation_recipe.spin import Nearest_Neighbour_1d
from quantum_simulation_recipe.trotter import pf, expH
from quantum_simulation_recipe.bounds import tight_bound
from evolution import Evolve_state
from states import zero_state

fig_dir, data_dir = './figs', './data'
```

## Example Usage

```python
model = 'nnh'
pf_order = 2
# n = 6
# p = 9e-3
# t, r = 5, 100
# t, r = 5, 30

n_list = [4, 5, 6, 7, 8, 9]
# n_list = [4, 5, 6]
# n_list = [10]
gamma_list = [2e-3, 3e-3, 4e-3, 5e-3, 6e-3, 7e-3, 8e-3, 9e-3, 10e-3]
# gamma_list = [2e-3]

if model == 'nnh':
    J, h, pbc = 2, 1, True
    model_par = {'Jz': J, 'hx': h, 'pbc': pbc}

r = 100
for gamma in gamma_list:
    for n in n_list:
        t = n
        dt = t/r
        print(f'n = {n}, t={t}, gamma = {gamma}')

        H_list = []
        nnh = Nearest_Neighbour_1d(n, Jz=J, hx=h, pbc=pbc)
        H_list = [term.to_matrix() for term in nnh.ham_xyz]


        U_dt_appro = pf(H_list, dt, 1)
        U_dt_exact = expH(sum(H_list), dt )

        input_state = zero_state(n)
        state_data = Evolve_state(U_dt_appro, U_dt_exact, r, input_state, gamma, pf_order, t)
        state_data.save(f'data/nnh_n/nnh_n{n}t{t}r{r}p{int(gamma*1e3)}.npy', savetxt=True)


# U_holistic = pf(H_list, t, r)
# U_holistic_exact = expH( sum(H_list), t )

# naive_alg_bnd = tight_bound(H_list, 2, dt, 1)
```

## Plot Fitting

```python
# Example usage

t, r = 5, 100
# n_list = [4, 5, 6, 7, 8, 9]
n_list = [4, 5, 6, 7, 8, 9, 10]
# n_list = [10]
p_list = [2e-3, 3e-3, 4e-3, 5e-3, 6e-3, 7e-3, 8e-3, 9e-3, 10e-3]
data_ = {}

from fitting import ReadVal, PolyFitting, AvgPolyFitting

for p in p_list:
    for n in n_list:
        # t = n
        print(f'n={n}, t={t}, p={p}')
        filedata = ReadVal(f'data/nnh_n/nnh_n{n}t{t}r{r}p{int(p*1e3)}.npy')
        data_[(n, p)] = filedata.data 

fittedval = {}

for p in p_list:
    n_files = { n: data_[(n, p)] for n in n_list }
    fitting_ = PolyFitting(r, t, p, n_files)
    fittedval[p] = fitting_

# Avg over p

avg_fittedval = AvgPolyFitting(fittedval)
```
